```python
# Data manipulation and visualization
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


# Standard libraries
import warnings
warnings.filterwarnings('ignore')
```

```python
# Function F(x, y)
def f(x, y):
    return (-y, x)
```

```python
def getVectorField(f, x_min, x_max, y_min, y_max,
                      x_step, y_step, unitary_field = False, stream_lines = False):

    # Meshgrid 
    x, y = np.meshgrid(np.linspace(x_min, x_max, x_step),  
                    np.linspace(y_min, y_max, y_step))
    
    # Get vector field elements
    u, v = f(x, y)

    # Plot
    fig, ax = plt.subplots()

    # Plotting vector field with Streamplot 
    if stream_lines:
        ax.streamplot(x, y, u, v, color='#B8BEDD')

    # If unitary field
    if unitary_field:
        pass

    # Plotting vector field with Quiver 
    ax.quiver(x, y, u, v, color = '#A11463', angles = 'xy')

    # Setting plot title 
    plt.title('Vector Field')

    # Setting x, y boundary limits 
    plt.xlim(x_min, x_max) 
    plt.ylim(y_min, y_max) 
    
    # Show plot with grid 
    plt.grid() 
    plt.show()
```

```python
# Test
getVectorField(f, x_min = -10, x_max = 10, y_min = -10, y_max = 10,
                x_step = 20, y_step = 20, stream_lines = True)
```
